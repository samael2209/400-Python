## Ejercicio 001: Entrada de usuario y strings

### Descripción
Pedir un nombre por teclado y saludar usando `input()` y f-strings.

### Conceptos aprendidos
- `input()`: Función para obtener datos del usuario
- `print()`: Mostrar información en pantalla

## Ejercicio 002: Operaciones aritméticas básicas

### Descripción
Solicitar dos números y mostrar suma, resta, multiplicación y división.

### Conceptos aprendidos
- `int()`: Convertir texto a número entero
- Operadores: `+`, `-`, `*`, `/`
- Condicional `if` para evitar división por cero

```python
# Código del ejercicio
num1 = int(input("Ingresa tu primer número: "))
num2 = int(input("Ingresa tu segundo número: "))

print(f"La suma de {num1} y {num2} es {num1 + num2}")
print(f"La resta de {num1} y {num2} es {num1 - num2}")
print(f"La multiplicación de {num1} y {num2} es {num1 * num2}")

if num2 != 0:
    print(f"La división de {num1} y {num2} es {num1 / num2}")
else:
    print("No se puede dividir entre cero")
```

## Ejercicio 003: Conversión de tipos de datos

### Descripción
Solicitar una edad y evaluar si es mayor de edad.

### Conceptos aprendidos
- `int()` para convertir input a entero
- Variables y constantes (mayoria_edad = 18)
- Operadores de comparación: `<=`

```python
# Código del ejercicio
edad = int(input("Ingresa tu edad: "))
mayoria_edad = 18

if edad <= mayoria_edad:
    print("Es menor")
else:
    print("Es mayor")
```

## Ejercicio 005: Estructura match-case (Python 3.10+)

### Descripción
Evaluar un color y mostrar mensaje según el caso.

### Conceptos aprendidos
- `.lower()`: Convertir texto a minúsculas
- `match-case`: Nueva estructura de control en Python 3.10
- `case _`: Caso default/por defecto

```python
# Código del ejercicio
color = input("Ingresa un color: ").lower()

match color:
    case 'verde':
        print("Puede pasar")
    case 'amarillo':
        print("Precaución al pasar")
    case 'rojo':
        print("No puede pasar")
    case _:
        print("Color inválido")
```

## Ejercicio 006: Ciclo for con range

### Descripción
Contar e imprimir números del 1 al 100.

### Conceptos aprendidos
- `range(inicio, fin)`: Generar secuencia de números
- Ciclo `for`: Iterar sobre una secuencia
- `range(1, 101)` genera números del 1 al 100

```python
# Código del ejercicio
for i in range(1, 101):
    print(i)
```

## Ejercicio 007: Números pares

### Descripción
Imprimir solo los números pares del 1 al 100.

### Conceptos aprendidos
- Operador módulo `%`: Obtiene el residuo de una división
- `i % 2 == 0`: Verifica si un número es par

```python
# Código del ejercicio
for i in range(1, 101):
    if i % 2 == 0:
        print(f"{i} es par")
```

## Ejercicio 008: Divisibles por 2 y 5

### Descripción
Clasificar números del 1 al 100 según sean divisibles por 2, 5 o ambos.

### Conceptos aprendidos
- Operadores lógicos: `and`
- Múltiples condiciones en un `if`
- `elif`: Segunda condición cuando la primera es falsa

```python
# Código del ejercicio
for i in range(1, 101):
    if i % 5 == 0 and i % 2 == 0:
        print(f"{i} es divisible por 2 y 5")
    elif i % 2 == 0:
        print(f"{i} es un número par")
    elif i % 5 == 0:
        print(f"{i} es divisible por 5")
    else:
        print(i)
```

## Ejercicio 009: Tablas de multiplicar

### Descripción
Mostrar las tablas de multiplicar del 2 al 9.

### Conceptos aprendidos
- Ciclos anidados: `for` dentro de `for`
- Rango con solo límite superior: `range(1, 10)`

```python
# Código del ejercicio
for i in range(1, 10):
    print(f"La tabla de {i}")
    for j in range(1, 10):
        print(f"{i} X {j} = {i * j}")
    print("-----")
```

## Ejercicio 010: Patrón de asteriscos (cuadrado)

### Descripción
Imprimir un cuadrado de 10x10 con asteriscos.

### Conceptos aprendidos
- `print(*, end=" ")`: Imprimir sin salto de línea
- `print()` vacío: Saltar línea
- Ciclos anidados para matrices

```python
# Código del ejercicio
for i in range(10):
    for j in range(10):
        print("*", end="  ")
    print()
```

## Ejercicio 011: Patrón de cuadro con bordes

### Descripción
Imprimir un rectángulo con bordes de asteriscos y espacio interior.

### Conceptos aprendidos
- Condicionales dentro de ciclos
- `print("*", " " * 26, "*", sep="")`: Imprimir con separador vacío

```python
# Código del ejercicio
for i in range(5):
    if i == 0 or i == 4:
        print("*  " * 10)
    else:
        print("*", " " * 26, "*", sep="")
```

## Ejercicio 012: Triángulo creciente

### Descripción
Imprimir un triángulo de asteriscos que crece hacia la derecha.

### Conceptos aprendidos
- `range(1, 6)`: Generar secuencia 1, 2, 3, 4, 5
- Multiplicación de strings: `"*" * 3` = `"***"`

```python
# Código del ejercicio
for i in range(1, 6):
    print("*" * i)
```

## Ejercicio 013: Triángulo decreciente

### Descripción
Imprimir un triángulo de asteriscos que decrece hacia la derecha.

### Conceptos aprendidos
- `range(inicio, fin, paso)`: Secencia con paso negativo
- `range(5, 0, -1)`: Genera 5, 4, 3, 2, 1

```python
# Código del ejercicio
for i in range(5, 0, -1):
    print(" * " * i)
```
