## Ejercicio 014: Función con texto centrado

### Descripción
Crear una función que escriba texto centrado en pantalla (80 columnas) con subrayado.

### Conceptos aprendidos
- Definición de funciones con `def`
- Parámetros de función
- Operaciones con strings y formato

```python
# Código del ejercicio
def EscribirCentrado(txt: str):
    a = 40 - len(txt) // 2
    print(" " * a + txt)
    print(" " * a + "=" * len(txt))

EscribirCentrado("¡Hola, Python!")
```

## Ejercicio 015: Verificar múltiplos

### Descripción
Crear una función que determine si un número es múltiplo de otro.

### Conceptos aprendidos
- Retorno de valores con `return`
- Tipado de parámetros y retorno
- Operador módulo `%`

```python
# Código del ejercicio
def EsMultiplo(n1: int, n2: int) -> bool:
    if n2 == 0:
        return False
    return n1 % n2 == 0

print(EsMultiplo(10, 5))  # True
print(EsMultiplo(7, 3))    # False
```

## Ejercicio 016: Temperatura media

### Descripción
Crear una función que calcule la temperatura media a partir de máximo y mínimo.

### Conceptos aprendidos
- Intercambio de variables
- Funciones con múltiples llamadas
- Bucles `for` con `range()`

```python
# Código del ejercicio
def TemperaturaMedia(temMax: float, temMin: float) -> float:
    if temMax < temMin:
        temMax, temMin = temMin, temMax
    return (temMax + temMin) / 2

print(TemperaturaMedia(25, 15))  # 20.0
```

## Ejercicio 017: Convertir espaciado

### Descripción
Crear una función que añada un espacio después de cada carácter.

### Conceptos aprendidos
- Iteración sobre strings
- Método `join()` para construir cadenas
- Comprehensions en strings

```python
# Código del ejercicio
def ConvertirEspaciado(txt: str) -> str:
    return "".join(c + " " for c in txt)

print(ConvertirEspaciado("Hola"))  # 'H o l a '
```

## Ejercicio 018: Máximo y mínimo de lista

### Descripción
Crear una función que devuelva el valor máximo y mínimo de una lista.

### Conceptos aprendidos
- Iteración sobre listas
- Retorno de tuplas
- Asignación múltiple

```python
# Código del ejercicio
def CalcularMaxMin(lista) -> tuple:
    maximo = minimo = lista[0]
    for num in lista:
        if num > maximo: maximo = num
        if num < minimo: minimo = num
    return maximo, minimo

nums = [3, 1, 9, 5, 2]
print(CalcularMaxMin(nums))  # (9, 1)
```

## Ejercicio 019: Circunferencia

### Descripción
Crear una función que calcule el área y perímetro de una circunferencia.

### Conceptos aprendidos
- `from math import pi`
- Funciones matemáticas
- Retorno de múltiples valores

```python
from math import pi
def Circunferencia(r: float) -> tuple:
    return pi * r**2, 2 * pi * r

area, perim = Circunferencia(5)
print(f"Área: {area:.2f}, Perímetro: {perim:.2f}")
```

## Ejercicio 020: Sistema de Login

### Descripción
Crear un sistema de login con 3 intentos máximos.

### Conceptos aprendidos
- While loops
- Contadores
- Lógica de validación

```python
def Login(usuario: str, password: str) -> bool:
    return usuario == "admin" and password == "1234"

# Ejemplo de uso
intentos = 3
while intentos > 0:
    if Login("admin", "1234"):
        print("¡Bienvenido!")
        break
    intentos -= 1
    print(f"Intentos restantes: {intentos}")
```

## Ejercicio 021: Factorial recursivo

### Descripción
Crear una función recursiva que calcule el factorial.

### Conceptos aprendidos
- Funciones recursivas
- Caso base
- Llamadas recursivas

```python
def Factorial(n: int) -> int:
    if n <= 1: return 1
    return n * Factorial(n - 1)

print(Factorial(5))  # 120
```

## Ejercicio 022: MCD (Método de Euclides)

### Descripción
Calcular el Máximo Común Divisor por el método de Euclides (recursivo).

### Conceptos aprendidos
- Recursividad
- Operador módulo
- Algoritmo de Euclides

```python
def MCD(a: int, b: int) -> int:
    if b == 0: return a
    return MCD(b, a % b)

print(MCD(48, 18))  # 6
```

## Ejercicio 023: Conversión de tiempo

### Descripción
Convertir entre segundos y horas/minutos/segundos.

### Conceptos aprendidos
- División entera //
- Módulo %
- match-case (Python 3.10+)

```python
def a_segundos(h, m, s): return h*3600 + m*60 + s
def a_hms(seg): return seg//3600, (seg%3600)//60, seg%60

print(a_segundos(1, 30, 0))    # 5400
print(a_hms(5400))            # (1, 30, 0)
```

## Ejercicio 024: Día juliano

### Descripción
Calcular el día juliano de una fecha.

### Conceptos aprendidos
- Listas como arrays
- Funciones auxiliaries
- Años bisiestos

```python
def es_bisiesto(año): return año%4==0 and (año%100!=0 or año%400==0)
def dias_mes(mes, año):
    dias = [31,28,31,30,31,30,31,31,30,31,30,31]
    return dias[mes-1] + (1 if mes==2 and es_bisiesto(año) else 0)
def dia_juliano(d, m, a):
    return sum(dias_mes(month, a) for month in range(1, m)) + d

print(dia_juliano(15, 3, 2024))  # 75 (bisiesto)
```

## Ejercicio 025: Validar fecha

### Descripción
Validar que una fecha sea correcta.

### Conceptos aprendidos
- Validación de entrada
- Funciones booleanas
- Condicionales

```python
def validar_fecha(d, m, a):
    if m < 1 or m > 12: return False
    if d < 1 or d > dias_mes(m, a): return False
    return True

print(validar_fecha(29, 2, 2024))  # True (bisiesto)
print(validar_fecha(29, 2, 2023))  # False
```

## Ejercicio 026: Fracciones

### Descripción
Trabajar con fracciones numerador/denominador.

### Conceptos aprendidos
- Manejo de excepciones
- raise ValueError
- Tuplas

```python
def fraccion(n, d):
    if d == 0: raise ValueError("Denominador no puede ser 0")
    return n, d, n/d

try:
    print(fraccion(1, 2))
except ValueError as e:
    print(e)
```

## Ejercicio 028: Máximo de tres

### Descripción
Devolver el mayor de tres números.

### Conceptos aprendidos
- Comparaciones encadenadas
- Lógica

```python
def max_de_tres(a, b, c):
    mayor = a
    if b >= mayor: mayor = b
    if c >= mayor: mayor = c
    return mayor

print(max_de_tres(3, 7, 5))  # 7
```

## Ejercicio 029: Longitud personalizada

### Descripción
Calcular la longitud de una lista o cadena (sin len()).

### Conceptos aprendidos
- Iteración manual
- Contadores

```python
def longitud(seq):
    cont = 0
    for _ in seq: cont += 1
    return cont

print(longitud("hola"))      # 4
print(longitud([1,2,3,4]))  # 4
```

## Ejercicio 031: Sum() y Multip()

### Descripción
Sumar y multiplicar todos los elementos de una lista.

### Conceptos aprendidos
- Acumuladores
- Iteración

```python
def suma(lista):
    total = 0
    for n in lista: total += n
    return total
def multip(lista):
    total = 1
    for n in lista: total *= n
    return total

nums = [1,2,3,4]
print(suma(nums))    # 10
print(multip(nums))  # 24
```

## Ejercicio 032: Invertir cadena

### Descripción
Invertir una cadena de texto.

### Conceptos aprendidos
- Iteración invertida
- Concatenación

```python
def inversa(txt):
    result = ""
    for c in txt: result = c + result
    return result

print(inversa("estoy"))  # 'yotse'
```

## Ejercicio 034: Superposición de listas

### Descripción
Verificar si dos listas tienen elementos en común.

### Conceptos aprendidos
- Sets y operaciones de conjuntos
- Bucles anidados

```python
def superposicion(l1, l2): return bool(set(l1) & set(l2))

print(superposicion([1,2,3], [4,5]))    # False
print(superposicion([1,2,3], [3,4]))    # True
```
