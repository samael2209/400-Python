## Ejercicio 037: Listas - Operaciones completas

### Descripción
A) Ingresar números a una lista (termina en 0)
B) Eliminar primera ocurrencia de un número
C) Calcular sumatoria de elementos
D) Filtrar números menores que X
E) Crear tuplas (número, frecuencia)

### Conceptos aprendidos
- Listas y métodos `append()`, `remove()`
- Iteración y condicionales
- Listas por comprensión
- Tuplas

```python
# Código del ejercicio
numeros = [1, 2, 3, 2, 4, 2]

# B) Eliminar primera ocurrencia
numeros.remove(2)
print(numeros)  # [1, 3, 2, 4, 2]

# C) Sumatoria
print(sum(numeros))  # 12

# D) Menores que 3
print([n for n in numeros if n < 3])  # [1, 2, 2]

# E) Frecuencia
from collections import Counter
print(Counter(numeros))
```

## Ejercicio 038: Pasajeros y Ciudades

### Descripción
Gestionar pasajeros con tuplas (nombre, DNI, destino) y ciudades con (ciudad, país).

### Conceptos aprendidos
- Listas de tuplas
- Búsqueda por clave
- Match-case

```python
pasajeros = [("Manuel", 12345, "Buenos Aires")]
paises = [("Buenos Aires", "Argentina")]

def buscar_por_dni(dni):
    for nombre, d, destino in pasajeros:
        if d == dni: return destino
    return "No encontrado"

print(buscar_por_dni(12345))
```

## Ejercicio 039: Conjuntos - Primaria vs Secundaria

### Descripción
Comparar listas de alumnos usando conjuntos.

### Conceptos aprendidos
- Sets (conjuntos)
- Intersección `&`
- Diferencia `-`

```python
primaria = {"Ana", "Juan", "Pedro"}
secundaria = {"Juan", "María", "Luis"}

print("Repetidos:", primaria & secundaria)     # {'Juan'}
print("Solo primaria:", primaria - secundaria) # {'Ana', 'Pedro'}
```

## Ejercicio 040:Diccionarios - Facturación

### Descripción
Obtener domicilios únicos de clientes para enviar facturas.

### Conceptos aprendidos
- Diccionarios
- Iteración sobre tuplas
- Acumulación de valores

```python
ventas = [
    ("Ana", 100, "Calle 1"),
    ("Ana", 200, "Calle 1"),
    ("Juan", 150, "Calle 2")
]

facturas = {}
for cliente, monto, dom in ventas:
    if cliente in facturas:
        facturas[cliente] += monto
    else:
        facturas[cliente] = monto

print(facturas)  # {'Ana': 300, 'Juan': 150}
```

## Ejercicio 041: Contador de caracteres

### Descripción
Contar ocurrencias de cada carácter en un string.

### Conceptos aprendidos
- Diccionarios para conteo
- Iteración sobre strings
- Método `.get()`

```python
texto = "hola mundo"
contador = {}
for c in texto:
    contador[c] = contador.get(c, 0) + 1
print(contador)
```

## Ejercicio 042: Club de Socios

### Descripción
Gestionar socios con diccionario: número → [nombre, fecha, cuota].

### Conceptos aprendidos
- Diccionarios con valores lista
- Modificación de diccionarios
- Eliminar elementos

```python
socios = {
    1: ["Ana Núñez", "17/03/2009", "s"],
    2: ["Juan Pérez", "20/05/2020", "n"]
}

# Agregar socio
socios[3] = ["María Gómez", "01/01/2024", "s"]

# Modificar
socios[2][2] = "s"  # pagó cuota

# Eliminar
del socios[1]

print(socios)
```

## Ejercicio 043: Contar Elementos

### Descripción
Contar cuántas veces aparece cada elemento en una lista.

### Conceptos aprendidos
- Diccionarios para conteo
- Iteración

```python
elementos = [1, 1, 2, 3, 2, 1, 4]
contador = {}
for e in elementos:
    contador[e] = contador.get(e, 0) + 1
print(contador)  # {1: 3, 2: 2, 3: 1, 4: 1}
```

## Ejercicio 044: Calcular Promedio

### Descripción
Calcular el promedio de calificaciones de estudiantes.

### Conceptos aprendidos
- `.values()`
- `sum()` y `len()`

```python
notas = {'Ana': 8, 'Juan': 6, 'María': 9}
promedio = sum(notas.values()) / len(notas)
print(f"Promedio: {promedio:.2f}")
```

## Ejercicio 045: Filtrar por Valor

### Descripción
Filtrar estudiantes approvalos de un diccionario.

### Conceptos/learned
- Dictionary comprehension
- Filtering

```python
estudiantes = {'Ana': 'aprobado', 'Juan': 'reprobado', 'María': 'aprobado'}
aprovados = {k: v for k, v in estudiantes.items() if v == 'aprobado'}
print(aprovados)
```

## Ejercicio 047: Contar Palabras

### Descripción
Contar cuántas veces aparece cada palabra en un texto.

### Conceptos aprendidos
- `.split()`
- Conteo con diccionarios

```python
texto = "hola hola mundo hola"
contador = {}
for palabra in texto.split():
    contador[palabra] = contador.get(palabra, 0) + 1
print(contador)
```

## Ejercicio 049: Valor Mínimo y Máximo

### Descripción
Encontrar valor min y max en diccionario.

### Conceptos aprendidos
- `values()`
- Iteración

```python
numeros = {'a': 10, 'b': 5, 'c': 20}
valores = list(numeros.values())
print(f"Min: {min(valores)}, Max: {max(valores)}")
```

## Ejercicio 050: Eliminar Claves

### Descripción
Eliminar claves si su valor > umbral.

### Conceptos/learned
- Filtering with comprehension

```python
d = {'a': 5, 'b': 20, 'c': 15}
umbral = 10
filtrado = {k: v for k, v in d.items() if v <= umbral}
print(filtrado)
```
